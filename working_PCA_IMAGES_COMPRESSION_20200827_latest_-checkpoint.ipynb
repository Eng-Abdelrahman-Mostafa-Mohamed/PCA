# PCA Function to Compress Gray Scale images

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler , OneHotEncoder
import numpy as np
import os
import cv2
from sklearn.datasets import fetch_olivetti_faces
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_compressing_PCA_GRAY_SCAL_IMAGES(image_path , num_components=56 ):

    Height = 224
    Width = 224
    image = cv2.imread(image_path, 0)
    image = cv2.resize(image, (Height, Width))

    # Center the data (subtract the mean)
    image_mean = np.mean(image)
    scaled_image = image - image_mean

    # Calculate the covariance matrix between features (pixels)
    cov_matrix_image = np.cov(scaled_image, rowvar=False)

    # Calculate eigenvalues and eigenvectors
    eigen_values, eigen_vectors = np.linalg.eig(cov_matrix_image)

    # Sort eigenvalues and their corresponding eigenvectors
    eigen_pairs = [(eigen_values[i], eigen_vectors[:, i]) for i in range(len(eigen_values))]
    eigen_pairs.sort(key=lambda x: x[0], reverse=True)

    # Select a subset of eigenvectors (number of components to keep)
    selected_eigenvectors = np.array([eigen_pairs[i][1] for i in range(num_components)]).T

    # Project the data onto the selected eigenvectors to compress the image
    compressed_image = np.dot(scaled_image, selected_eigenvectors)

    # Reconstruct the compressed image
    reconstructed_image = np.dot(compressed_image, selected_eigenvectors.T) + image_mean
    reconstructed_image = reconstructed_image.astype(np.uint8)

    # Display the original, compressed, and reconstructed images
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')

    plt.subplot(1, 3, 2)
    plt.title('Compressed Image')
    # Reshape the compressed_image based on its actual size
    compressed_height, compressed_width = compressed_image.shape
    plt.imshow(compressed_image.reshape(compressed_height, compressed_width), cmap='gray')

    plt.subplot(1, 3, 3)
    plt.title('Reconstructed Image')
    plt.imshow(reconstructed_image, cmap='gray')

    plt.show()

# Call the function with the image path
image_compressing_PCA_GRAY_SCAL_IMAGES('/content/test_image.jpeg')
```

# PCA RGB (COLORED IMGS) COMPRESSION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imageio
def image_compressing_PCA_COLOR_IMAGES(image_path,num_components = 20 ):
    Height = 408
    Width = 612
    # image = cv2.imread(image_path)
    image = cv2.imread(image_path)
    image = cv2.resize(image, (Height, Width))

    # Separate the color channels (R, G, B)
    red_channel = image[:, :, 0]
    green_channel = image[:, :, 1]
    blue_channel = image[:, :, 2]

    # Initialize arrays to store compressed and reconstructed channels
    compressed_red_channel = None
    compressed_green_channel = None
    compressed_blue_channel = None
    reconstructed_red_channel = None
    reconstructed_green_channel = None
    reconstructed_blue_channel =None

    # Process each channel separately
    for channel, channel_name in zip([red_channel, green_channel, blue_channel], ['Red', 'Green', 'Blue']):
        # Center the data (subtract the mean)
        channel_mean = np.mean(channel)
        scaled_channel = channel - channel_mean

        # Calculate the covariance matrix between features (pixels)
        cov_matrix_channel = np.cov(scaled_channel, rowvar=False)

        # Calculate eigenvalues and eigenvectors
        eigen_values, eigen_vectors = np.linalg.eig(cov_matrix_channel)

        # Sort eigenvalues and their corresponding eigenvectors
        eigen_pairs = [(eigen_values[i], eigen_vectors[:, i]) for i in range(len(eigen_values))]
        eigen_pairs.sort(key=lambda x: x[0], reverse=True)

        # Select a subset of eigenvectors (number of components to keep)

        selected_eigenvectors = np.array([eigen_pairs[i][1] for i in range(num_components)]).T

        # Project the data onto the selected eigenvectors to compress the channel
        compressed_channel = np.dot(scaled_channel, selected_eigenvectors)

        # Reconstruct the compressed channel
        reconstructed_channel = np.dot(compressed_channel, selected_eigenvectors.T) + channel_mean
        reconstructed_channel = reconstructed_channel.astype(np.uint8)

        # Store the compressed and reconstructed channels
        if channel_name == 'Red':
            compressed_red_channel = compressed_channel
            reconstructed_red_channel = reconstructed_channel
        elif channel_name == 'Green':
            compressed_green_channel = compressed_channel
            reconstructed_green_channel = reconstructed_channel
        else:
            compressed_blue_channel = compressed_channel
            reconstructed_blue_channel = reconstructed_channel

        # Display the original, compressed, and reconstructed channels
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.title(f'{channel_name} Channel (Original)')
        plt.imshow(channel)

        plt.subplot(1, 3, 2)
        plt.title(f'{channel_name} Channel (Compressed)')
        # Reshape the compressed_channel based on its actual size
        compressed_height, compressed_width = compressed_channel.shape
        plt.imshow(compressed_channel.reshape(compressed_height, compressed_width))

        plt.subplot(1, 3, 3)
        plt.title(f'{channel_name} Channel (Reconstructed)')
        plt.imshow(reconstructed_channel)

        plt.show()

    # Combine the compressed and reconstructed channels to get RGB images
    compressed_image = cv2.merge([compressed_red_channel,compressed_green_channel,compressed_blue_channel])
    reconstructed_image = cv2.merge([reconstructed_red_channel,reconstructed_green_channel,reconstructed_blue_channel])
    # Display the original, compressed, and reconstructed RGB images
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title('Original RGB Image')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 3, 2)
    plt.title('Compressed RGB Image')
    # Convert the reconstructed image to 8-bit format before displaying
    compressed_image_8bit = cv2.convertScaleAbs(compressed_image)
    plt.imshow(cv2.cvtColor(compressed_image_8bit, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 3, 3)
    plt.title('Reconstructed RGB Image')
    # Convert the reconstructed image to 8-bit format before displaying
    reconstructed_image_8bit = cv2.convertScaleAbs(reconstructed_image)
    plt.imshow(cv2.cvtColor(reconstructed_image_8bit, cv2.COLOR_BGR2RGB))

plt.show()



# Call the function with the image path of an RGB image
image_compressing_PCA_COLOR_IMAGES('/content/colored.jpg')
```

# Eigen Faces

compress a group of images into less number of images contains common Features


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces

def eigen_faces(images_data, num_components):
    # Calculate the mean image
    mean_image__ = np.mean(images_data, axis=0)

    # Convert the list of images to a NumPy array and subtract the mean
    list_scaled_images = images_data - mean_image__

    # Reshape the 3D data to 2D
    num_samples, h, w = list_scaled_images.shape
    list_scaled_images = list_scaled_images.reshape(num_samples, h * w)

    # Calculate the covariance matrix
    cov_matrix_all_data = np.cov(list_scaled_images)

    eigen_values, eigen_vectors = np.linalg.eig(cov_matrix_all_data)
    print(f'the shape of eigen values is {eigen_values.shape}')
    print(f'the shabe of eigen vector{eigen_vectors.shape}')
    eigen_pairs = [(eigen_values[i], eigen_vectors[:, i]) for i in range(len(eigen_values))]
    eigen_pairs.sort(key=lambda x: x[0], reverse=True)

    selected_eigenvectors = np.array([eigen_pairs[i][1] for i in range(num_components)]).T
    eigenfaces = np.dot(selected_eigenvectors.T, list_scaled_images) # Changed this line
    print(f'the shape of eigen faces is --->{eigenfaces.shape}')
    plt.figure(figsize=(15, 5))
    for i in range(num_components):
        plt.subplot(1, num_components, i + 1)
        plt.title(f'Eigenface {i+1}')
        eigenface = eigenfaces[i].reshape(64, 64) # Changed this line
        plt.imshow(eigenface, cmap='gray')

    plt.show()

# Load the Olivetti Faces dataset
dataset = fetch_olivetti_faces(shuffle=True, random_state=0)
images = dataset.images

# Pass the reshaped images array as the first argument to your eigen_faces function
eigen_faces(images, num_components=8)
```
